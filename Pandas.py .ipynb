# Aggregation functions

```python
# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Applying multiple aggregation functions
result = df.agg({
    'A': ['sum', 'mean','min', 'max'],
    'B': ['sum', 'mean','min', 'max']
})

print(result)
```

```python
# Changing data types
df['A'] = df['A'].astype(float)
print(df)
print(df.dtypes)
```

```python
# Filter values between 2 and 4
filtered = df[df['A'].between(2, 4)]
print(filtered)
```

```python
# Clip values below 2 and above 4
clipped = df['A'].clip(lower=2, upper=4)
print(clipped)
```

```python
# Create DataFrame with duplicates
df_with_dupes = pd.DataFrame({
    'A': [1, 1, 2, 2, 3],
    'B': [10, 10, 20, 20, 30]
})

# Drop duplicates
df_no_dupes = df_with_dupes.drop_duplicates()
print(df_no_dupes)
```

```python
# Create DataFrame with NaNs
df_with_nans = pd.DataFrame({
    'A': [1, 2, None, 4, 5],
    'B': [10, None, 30, 40, 50]
})

# Drop rows with NaNs
df_no_nans = df_with_nans.dropna()
print(df_no_nans)
```

```python
# Fill NaNs with a specified value
df_filled = df_with_nans.fillna(0)
print(df_filled)
```

```python
# Filter columns by labels
df_filtered = df.filter(['A'])
print(df_filtered)
```

```python
# Select rows and columns by position
selected = df.iloc[0:2, 0:1]
print(selected)
```

```python
# Detect missing values
print(df_with_nans.isnull())
```

```python
# Map values in a Series
s = pd.Series(['cat', 'dog', 'cat'])
mapped = s.map({'cat': 'kitten', 'dog': 'puppy'})
print(mapped)
```

```python
# Filter rows by axis=0 (rows) and criteria
df_filtered_rows = df[df['A'] > 2]
print(df_filtered_rows)
```
