```python
# Read the CSV file
df=pd.read_csv(r"E:\datasets\datasets-session-23\batter.csv")

# Display the first few rows of the dataframe to confirm it was read correctly
print(df.head())
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title('AVG VS SR')
plt.xlabel("average")
plt.ylabel("strike rate")
plt.show()
```

```python
# datasets of top 50 batter y runs 
df=df.sort_values('runs',ascending=False).head(50)
plt.scatter(df['avg'],df['strike_rate'],color='red',s=df['runs']/10)
plt.title('AVG VS SR')
plt.xlabel("average")
plt.ylabel("strike rate")
plt.show()
```

```python
# Assuming df is already defined and contains the 'batter', 'runs', 'avg', and 'strike_rate' columns
df = df.sort_values('runs', ascending=False).head(20)

# Create a color map
colors = plt.cm.get_cmap('tab20', len(df))

# Plot each batter with a unique color
plt.figure(figsize=(12, 8))
for i, batter in enumerate(df['batter'].unique()):
    batter_data = df[df['batter'] == batter]
    plt.scatter(batter_data['avg'], batter_data['strike_rate'], color=colors(i), s=batter_data['runs'] / 10, label=batter)

plt.title('AVG VS SR')
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.legend(title='Batter', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# surface plots

```python
x= np.linspace(-10,10,100)
y= np.linspace(-10,10,100)

#  for grid
xx ,yy = np.meshgrid(x,y)

z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
